```python
def create_base_state(story_content: str = SAMPLE_STORY):
    """ê¸°ë³¸ í…ŒìŠ¤íŠ¸ ìƒíƒœ ìƒì„±"""
    return {
        "content": story_content,
        "project_id": "test-project-001",
        "document_id": "test-document-001",
        "job_id": "test-job-001",
        "callback_url": "http://localhost:8080/api/callback",
        "existing_characters": [],
        "existing_events": [],
        "existing_relationships": [],
        "extracted_characters": [],
        "extracted_events": [],
        "extracted_settings": {},
        "analyzed_dialogues": {},
        "tracked_emotions": {},
        "relationship_graph": {},
        "consistency_report": {},
        "plot_integration": {},
        "validation_result": {},
        "extraction_done": False,
        "analysis_done": False,
        "validation_done": False,
        "retry_count": 0,
        "messages": [],
        "errors": []
    }

def pretty_print_result(title: str, result: dict):
    """ê²°ê³¼ë¥¼ ë³´ê¸° ì¢‹ê²Œ ì¶œë ¥"""
    print(f"\n{'='*60}")
    print(f"ğŸ“‹ {title}")
    print(f"{'='*60}")
    print(json.dumps(result, ensure_ascii=False, indent=2))
    print(f"{'='*60}\n")

# Jupyterì—ì„œ async í•¨ìˆ˜ ì‹¤í–‰ì„ ìœ„í•œ í—¬í¼
def run_async(coro):
    """ë¹„ë™ê¸° í•¨ìˆ˜ ì‹¤í–‰"""
    try:
        loop = asyncio.get_event_loop()
        if loop.is_running():
            import nest_asyncio
            nest_asyncio.apply()
            return loop.run_until_complete(coro)
        return asyncio.run(coro)
    except RuntimeError:
        return asyncio.run(coro)

print("í—¬í¼ í•¨ìˆ˜ ì¤€ë¹„ ì™„ë£Œ!")
```
